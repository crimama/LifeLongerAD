```python
import pandas as pd 
import numpy as np 
import os 
from glob import glob 
import matplotlib.pyplot as plt 
```

```python
data_dir = '/Volume/VAD/LifeLongerAD/results/LANGCAD/MVTecAD/*/seed_0/result.csv'
result_dirs = glob(data_dir)

df = pd.DataFrame()
for rd in result_dirs:
    temp = pd.read_csv(rd)
    if 'task-agnostic' not in temp.columns:
        temp.loc[::2,'task-agnostic'] = 'specific'
        temp.loc[1::2,'task-agnostic'] = 'agnostic'    
        temp['exp_name'] = rd.split('/')[-3]
    df = pd.concat([df,temp])
    
```

```python
specific_df = df[df['task-agnostic'] == 'specific']

# 'num_blocks'별로 그룹화하고, 각 metric의 평균 계산
result = specific_df.groupby('exp_name')[
    ['img_level_auroc', 'img_level_average_precision', 'pix_level_auroc', 'pix_level_average_precision']
].mean()

# 결과 출력
result['average'] = result.mean(axis=1)
result = result.sort_values(by='average', ascending=False)
result
```
